```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
ip ="/content/realimage.jpg"
plt.imshow(Image.open(ip))
plt.axis('off')
```

```python
img = Image.open(ip)
text = pytesseract.image_to_string(img)
print("Extracted text:" ,text)
```

```python
audio_output = gTTS(text=text)
audio_output.save("extracted_text_audio.mp3")
Audio("extracted_text_audio.mp3")
```
