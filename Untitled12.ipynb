```python
from pydub import AudioSegment
import speech_recognition as sr
```

```python
audio = AudioSegment.from_file("/content/Meeting India Market.mp3")
audio.export("temp.wav", format="wav")
```

```python
r = sr.Recognizer()
with sr.AudioFile("temp.wav") as source:
    audio_data = r.record(source)
    try:
        text = r.recognize_google(audio_data)
        print("Text from audio:")
        print(text)
    except sr.UnknownValueError:
        print("Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
```

```python
from googletrans import Translator
summary=''' India market exploration date May 21st 2025 attendees leadership team objective
 assess viability of India market entry and Define next steps One Market potential preliminary
  research shows robust GDP growth and a burgeoning middle class .'''
languages = {
    '1': ('French', 'fr'),
    '2': ('Spanish', 'es'),
    '3': ('German', 'de'),
    '4': ('Japanese', 'ja'),
    '5': ('Hindi', 'hi'),
    '6': ('Tamil', 'ta'),
    '7': ('Telugu', 'te')
}
choice = input("Enter the number of your choice: ").strip()

if choice in languages:
    lang_name, lang_code = languages[choice]
    translator = Translator()
    translated = translator.translate(summary, dest=lang_code)
    print(f"\n Translated Summary in {lang_name}:")
    print(translated.text)
else:
    print(" Invalid choice.")
```
